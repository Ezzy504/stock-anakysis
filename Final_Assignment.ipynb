## Objective:
Extract stock data and revenue data for Tesla and GameStop using Python, and display them using interactive Plotly dashboards.


### Q1: Extracting Tesla Stock Data using `yfinance`


```python
import yfinance as yf
import pandas as pd

tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

### Q2: Extracting Tesla Revenue Data using Web Scraping


```python
import requests
from bs4 import BeautifulSoup

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")
```

```python
import lxml  # make sure lxml is installed
tables = pd.read_html(html_data)

tesla_revenue = tables[1]
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != '']
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(str)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(",", "")
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace("$", "")
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"], errors="coerce")

tesla_revenue.head()
```

### Q3: Extracting GameStop Stock Data using `yfinance`



```python
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
import pandas as pd

# Step 1: Read all tables from Macrotrends revenue page
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
tables = pd.read_html(url)

# Step 2: Check number of tables
print(f"Total tables found: {len(tables)}")

# Step 3: Preview each table to identify the correct one
for i, table in enumerate(tables):
    print(f"\nTable {i}")
    print(table.head())
# Extract the correct revenue table (adjust index if needed)
gme_revenue = tables[1]
gme_revenue.columns = ["Date", "Revenue"]

# Clean the Revenue column
gme_revenue["Revenue"] = gme_revenue["Revenue"].replace("[$,]", "", regex=True)
gme_revenue = gme_revenue[gme_revenue["Revenue"] != ""]
gme_revenue.dropna(inplace=True)
gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"])

# Convert Date column to datetime format
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"])

# Preview cleaned data
gme_revenue.head()
```

```python
import plotly.graph_objects as go

# Create a dashboard for Tesla Stock and Revenue
fig = go.Figure()

# Add Tesla stock price trace
fig.add_trace(go.Scatter(
    x=tesla_data["Date"],
    y=tesla_data["Close"],
    name="Tesla Stock Price",
    mode="lines"
))

# Add Tesla revenue trace
fig.add_trace(go.Scatter(
    x=tesla_revenue["Date"],
    y=tesla_revenue["Revenue"],
    name="Tesla Revenue",
    mode="lines"
))

# Update layout
fig.update_layout(
    title="Tesla Stock Price and Revenue Over Time",
    xaxis_title="Date",
    yaxis_title="USD",
    legend_title="Data Type",
    template="plotly_white"
)

fig.show()
```

```python
import plotly.graph_objects as go

# Ensure both DataFrames are loaded
# gme_data should have "Date" and "Close"
# gme_revenue should have "Date" and "Revenue"

# Convert gme_revenue["Date"] to datetime to match gme_data
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"])
gme_data["Date"] = pd.to_datetime(gme_data["Date"])

# Create the interactive figure
fig = go.Figure()

# Add GameStop stock price trace
fig.add_trace(go.Scatter(
    x=gme_data["Date"],
    y=gme_data["Close"],
    name="GameStop Stock Price",
    mode="lines",
    line=dict(color="blue")
))

# Add GameStop revenue trace
fig.add_trace(go.Scatter(
    x=gme_revenue["Date"],
    y=gme_revenue["Revenue"],
    name="GameStop Revenue",
    mode="lines",
    line=dict(color="green"),
    yaxis="y2"
))

# Update layout with dual y-axes
fig.update_layout(
    title="GameStop Stock Price and Quarterly Revenue Over Time",
    xaxis_title="Date",
    yaxis=dict(
        title="Stock Price (USD)",
        titlefont=dict(color="blue"),
        tickfont=dict(color="blue")
    ),
    yaxis2=dict(
        title="Revenue (USD in Millions)",
        titlefont=dict(color="green"),
        tickfont=dict(color="green"),
        overlaying="y",
        side="right"
    ),
    legend=dict(x=0.01, y=0.99),
    template="plotly_white"
)

# Show the dashboard
fig.show()

```
